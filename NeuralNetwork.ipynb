```python
import numpy as np
import pandas as pd 
from keras.models import Sequential
from keras.layers import *
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adam
from keras.utils import plot_model
```

```python
df=pd.read_csv('labelled_data.csv')
df.head(n=10)
```

```python
y=df['target'].values.reshape((-1,1))
x=df.iloc[:,-3:-1].values
```

```python
print(x.shape,y.shape)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8)
print(xtrain.shape,ytrain.shape,xtest.shape,ytest.shape)
```

```python
from keras.utils import to_categorical
ytrain=to_categorical(ytrain,num_classes=6)
ytest=to_categorical(ytest,num_classes=6)
```

# Model

```python
model=Sequential()
model.add(Dense(100,activation='relu',input_shape=(2,)))
model.add(Dropout(0.3))
model.add(BatchNormalization())
model.add(Dense(20,activation='relu'))
model.add(Dropout(0.3))
model.add(BatchNormalization())
model.add(Dense(20,activation='relu'))
model.add(Dropout(0.3))
model.add(BatchNormalization())
model.add(Dense(6,activation='sigmoid'))
model.summary()
```

```python
plot_model(model, show_shapes=True, to_file='model.png')
```

# Model Training

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
checkpoint=ModelCheckpoint('model.h5',monitor='val_accuracy',save_best_only=False,save_weights_only=False,
    mode='auto',save_freq='epoch',)
```

```python
hist=model.fit(xtrain,ytrain,batch_size=8,epochs=200,validation_data=(xtest,ytest),verbose=1,callbacks=[checkpoint])
```

```python
opt=Adam(learning_rate=0.0001)
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
hist2=model.fit(xtrain,ytrain,batch_size=8,epochs=200,validation_data=(xtest,ytest),verbose=1,callbacks=[checkpoint])
```

```python
val_acc=[x for x in hist.history['val_accuracy']]
for x in hist2.history['val_accuracy']:
    val_acc.append(x)
acc=[x for x in hist.history['accuracy']]
for x in hist2.history['accuracy']:
    acc.append(x)
```

```python
# plt.style.use('seaborn')
# plt.plot(hist.history['accuracy'],label='accuracy')
# plt.plot(hist.history['val_accuracy'],label='val_accuracy')
# plt.legend()
# plt.show()
plt.style.use('seaborn')
plt.plot(acc,label='accuracy')
plt.plot(val_acc,label='val_accuracy')
plt.legend()
plt.show()
```

```python
model.load_weights('model.h5')
```

```python
model.evaluate(xtest,ytest)
model.evaluate(xtrain,ytrain)
```
